```python
import numpy as np
import pandas as pd
```

```python
colum_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('u.data', sep='\t', names=colum_names)
df.head()
```

```python
movies_title = pd.read_csv('Movie_Id_Titles')
movies_title.head()
```

```python
df = pd.merge(df, movies_title, on='item_id')
```

```python
df.head()
```

```python
#Movies with the best ratings
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings['num of ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
movies =  df.pivot_table(index='user_id', columns='title', values='rating')
movies.head()
```

```python
ratings.sort_values('num of ratings', ascending=False).head(10) #first ten movies with the highest number of ratings
```

```python
fargo_user_ratings = movies['Fargo (1996)']
toystory_user_ratings = movies['Toy Story (1995)']
terminator_user_ratings = movies['Terminator, The (1984)']
backtofuture_user_ratings = movies['Back to the Future (1985)']
fargo_user_ratings.head()
```

```python
#Using corrwith() to get correlation between the 4 series
similar_to_fargo = movies.corrwith(fargo_user_ratings)
similar_to_toystory = movies.corrwith(toystory_user_ratings)
similar_to_terminator = movies.corrwith(terminator_user_ratings)
similar_to_backtofuture = movies.corrwith(backtofuture_user_ratings)
```

```python
#Removing NAN Values and finding similar movies for FARGO based on ratings.
corr_fargo = pd.DataFrame(similar_to_fargo, columns=['Correlation'])
corr_fargo.dropna(inplace=True)
corr_fargo = corr_fargo.join(ratings['num of ratings'])
corr_fargo[corr_fargo['num of ratings']>100].sort_values('Correlation', ascending=False).head(10)
```

#### These are the recomended movies for FARGO based on user ratings

```python
#Removing NAN Values and finding similar movies for TOY STORY based on ratings.
corr_toystory = pd.DataFrame(similar_to_toystory, columns=['Correlation'])
corr_toystory.dropna(inplace=True)
corr_toystory = corr_toystory.join(ratings['num of ratings'])
corr_toystory[corr_toystory['num of ratings']>100].sort_values('Correlation', ascending=False).head(10)
```

#### These are the recomended movies for TOY STORY based on user ratings

```python
#Removing NAN Values and finding similar movies for TERMINATOR based on ratings.
corr_terminator = pd.DataFrame(similar_to_terminator, columns=['Correlation'])
corr_terminator.dropna(inplace=True)
corr_terminator = corr_terminator.join(ratings['num of ratings'])
corr_terminator[corr_terminator['num of ratings']>100].sort_values('Correlation', ascending=False).head(10)
```

#### These are the recomended movies for TERMINATOR, THE based on user ratings

```python
#Removing NAN Values and finding similar movies for BACK TO THE FUTURE based on ratings.
corr_backtofuture = pd.DataFrame(similar_to_backtofuture, columns=['Correlation'])
corr_backtofuture.dropna(inplace=True)
corr_backtofuture = corr_backtofuture.join(ratings['num of ratings'])
corr_backtofuture[corr_backtofuture['num of ratings']>100].sort_values('Correlation', ascending=False).head(10)
```
